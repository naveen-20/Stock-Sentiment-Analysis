In the dataset we can see that there are headlines from 25 different newspaper. Let us join them together in one column. 

```python
" ".join(str(x) for x in data.iloc[1,2:])
```

```python
Headline = []
for i in range(0,len(data.index)):
    Headline.append(" ".join(str(x) for x in data.iloc[i,2:]))
```

```python
headlines = pd.DataFrame(Headline,columns=['Headline'])
```

```python
len(headlines)
```

```python
data = data.iloc[:,0:2]
```

```python
data.head()
```

```python
new_data = pd.concat([data,headlines],axis=1)
```

```python
new_data.head()
```

```python
new_data.tail()
```

Data Preprocessing.

```python
import re
import nltk
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
```

```python
wordnet = WordNetLemmatizer()
```

```python
corpus = []
for i in range(len(new_data)):
    message = re.sub('[^A-Za-z]',' ',new_data['Headline'][i])
    message = message.lower()
    message = message.split()
    message = [wordnet.lemmatize(word) for word in message if word not in stopwords.words('english')]
    message = " ".join(message)
    corpus.append(message)
```

```python
corpus[1:5]
```

```python
new_data['Headline'] = corpus
```

```python
new_data.head()
```

```python
new_data.tail()
```

Splitting the dataset.

```python
train = new_data[new_data['Date'] < '20100101']
test = new_data[new_data['Date'] > '20091231']
```

```python
train.head()
```

```python
test.head()
```

Creating bag of words.

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(ngram_range=(5,5))
```

```python
X_train = cv.fit_transform(train['Headline'])
```

```python
X_test = cv.transform(test['Headline'])
```

```python
print(X_train.shape)
print(X_test.shape)
```

```python
y_train = train['Label']
```

```python
y_test = test['Label']
```

```python
print(y_train.shape)
print(y_test.shape)
```

Implementing random forest classificaton.

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier(n_estimators=200,criterion='entropy')
rf.fit(X_train,y_train)
```

Making predictions.

```python
y_pred = rf.predict(X_test)
```

Checking the accuracy.

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
```

```python
matrix=confusion_matrix(y_test,y_pred)
print(matrix)
score=accuracy_score(y_test,y_pred)
print(score)
report=classification_report(y_test,y_pred)
print(report)
```
